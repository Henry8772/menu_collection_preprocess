```python
import json
import os
```

```python
# Load the JSON file, or create a new one if it does not exist
if os.path.exists('blue-frog-menu.json'):
    with open('blue-frog-menu.json', 'r', encoding='utf-8') as f:
        print('Loading existing JSON file...')
        restaurants = json.load(f)
```

```python
restaurants
```

```python
result_dict = {}

# Combine entries with the same name and ensure no duplicate photos
for entry in restaurants:
    name = entry['name']
    
    # If the name is not in the result_dict, add it
    if name not in result_dict:
        result_dict[name] = {
            "name": name,
            "link": entry["link"],
            "menu_photos": set(),
            "receipt_photos": set(),
            "mix_photos": set()
        }
    
    # Add the photos to the appropriate sets (this will automatically ensure uniqueness)
    result_dict[name]["menu_photos"].update(entry["menu_photos"])
    result_dict[name]["receipt_photos"].update(entry["receipt_photos"])
    result_dict[name]["mix_photos"].update(entry["mix_photos"])

# Convert sets back to lists and dictionary to list
result_list = []
for name, entry in result_dict.items():
    entry["menu_photos"] = list(entry["menu_photos"])
    entry["receipt_photos"] = list(entry["receipt_photos"])
    entry["mix_photos"] = list(entry["mix_photos"])
    result_list.append(entry)

```

```python
# Convert dictionary to list
result_list = list(result_dict.values())

# Convert to JSON
# result_json = json.dumps(result_list, indent=4, ensure_ascii=False)
# print(result_json)

# Save the updated JSON after each type of photos are processed
with open('blue-frog-menu.json', 'w', encoding='utf-8') as f:
    json.dump(result_list, f, ensure_ascii=False, indent=4)
```
